```python
import json
```

```python
# with open('src/solver/-5097.solution.json') as f:
#     data = json.load(f)

with open('doctors.-19014.solution.json') as f:
    data_doc = json.load(f)

with open('nurses.-1005.solution.json') as f:
    data_nurse = json.load(f)

with open('rtg_assistants.-18016.solution.json') as f:
    data_rtg = json.load(f)
```

```python
data = []
data.extend(json.loads(data_doc["solution"]))
data.extend(json.loads(data_nurse["solution"]))
data.extend(json.loads(data_rtg["solution"]))
```

```python
solution = data
```

```python
names = set()
shift_types = set()
dates = set()

for item in solution:
    dates.add(item['shift']["datetype"]["date"])
    shift_types.add(item['shift']["datetype"]["type"])
    names.add(item["employee"])
    # item["department"]

date_index_map = {dt: i for i, dt in list(enumerate(sorted(list(dates))))}
```

```python
solution[0]
```

```python
shift_type_map = {
    "MORNING": "Ranní",
    "AFTERNOON": "Odpo",
    "WEEKEND_MORNING": "Ranní 12",
    "WEEKEND_EVENING": "Odpo 12",
    "MORNING_12": "Ranní 12",
    "EVENING_12": "Odpo 12",
    "OVERNIGHT": "Noční",
}
```

```python
transformed_solution = []
for item in solution:
    new_item = {
        "date": item['shift']["datetype"]["date"],
        "date_index": date_index_map[item['shift']["datetype"]["date"]],
        "part_index": item["shift"]["department"] + " " + shift_type_map[item['shift']["datetype"]["type"]],
        "name": item["employee"],
        "shift_type": shift_type_map[item['shift']["datetype"]["type"]],
        "occupation": item["employee"].split(" ")[0],
        "department": item["shift"]["department"],
        "preference": item["preference"]
    }
    transformed_solution.append(new_item)
```

```python
parts = set()
date_index = set()

for item in transformed_solution:
    parts.add(item["part_index"])
    date_index.add(item["date_index"])

parts = sorted(list(parts))
date_index = sorted(list(date_index))
```

```python
parts
```

```python
date_index
```

```python
final_json = {}
for p in parts:
    final_json[p] = {}
    for di in date_index:
        final_json[p][di] = []
        for item in transformed_solution:
            if item["date_index"] == di and item["part_index"] == p:
                final_json[p][di].append(item)
```

```python
with open("./regrouped_json.json", "w", encoding='utf-8') as fp:
    json.dump(final_json, fp, ensure_ascii=False)
```
